```python
import numpy as np
from sklearn.datasets import make_circles
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense
from keras.models import Sequential
from tensorflow.keras.optimizers import SGD, Adam, Adadelta, Adagrad ,RMSprop
from keras.initializers import RandomUniform ,RandomNormal, he_uniform, he_normal, glorot_normal , glorot_uniform
from matplotlib import pyplot as plt
```

```python
X,y=make_circles(n_samples=1000,noise=0.1,random_state=0)
```

```python
for i in range(2):
    samples_ix=np.where(y==i)
    plt.scatter(X[samples_ix,0], X[samples_ix,1], label=str(i))
plt.legend()
```

```python
scaler=MinMaxScaler()
X_trf=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X , train_y ,test_y=train_test_split(X_trf, y, test_size=0.33)
```

```python
train_X.shape, test_X.shape , train_y.shape ,test_y.shape
```

```python
model=Sequential()
model.add(Dense(5, activation='tanh',kernel_initializer=RandomUniform(minval=0, maxval=1), input_dim=2))
model.add(Dense(1, activation='sigmoid',kernel_initializer=RandomUniform(minval=0, maxval=1)))
```

```python
model.summary()
```

```python
opt=SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics='accuracy')
```

```python
history=model.fit(train_X, train_y, validation_data=(test_X, test_y), epochs=500, verbose=0 )
```

```python
history.history.keys()
```

```python
plt.figure(figsize=(10,4))
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
```

# Depper net with tanh

```python
model=Sequential()
initializer=RandomUniform(minval=0, maxval=1)
model.add(Dense(5, activation='tanh',kernel_initializer=initializer, input_dim=2))
model.add(Dense(5, activation='tanh',kernel_initializer=initializer))
model.add(Dense(5, activation='tanh',kernel_initializer=initializer))
model.add(Dense(5, activation='tanh',kernel_initializer=initializer))
model.add(Dense(5, activation='tanh',kernel_initializer=initializer))
model.add(Dense(1, activation='sigmoid',kernel_initializer=initializer))
```

```python
opt=SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics='accuracy')
```

```python
history=model.fit(train_X, train_y, validation_data=(test_X, test_y), epochs=500, verbose=0 )
```

```python
history.history.keys()
```

```python
plt.figure(figsize=(10,4))
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
```
