```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('pima-indians-diabetes.csv' , header= None)
```

```python
data.head()
```

```python
x=data.loc[:,:7]
y=data.loc[:,8]
y.head()
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2, random_state=30)
```

```python
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
model=Sequential()
model.add(Dense(32, activation='sigmoid', input_dim=8))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='SGD',metrics='accuracy')
```

```python
his=model.fit(train_x,train_y,validation_data=[test_x,test_y], epochs=500, batch_size=32,verbose=0)
```

```python
his.history['accuracy']
```

```python
plt.plot(his.history['accuracy'], label='train')
plt.plot(his.history['val_accuracy'], label='test')
plt.legend()
```
