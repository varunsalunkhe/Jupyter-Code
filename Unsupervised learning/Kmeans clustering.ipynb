```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns

x, y = make_blobs(n_samples=1000, centers=3, n_features=2,
                  random_state=42)
print(x.shape)
```

```python
sns.set_style('darkgrid')
plt.scatter(x[:,0] , x[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.23, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    k_cluster = KMeans(n_clusters=i, init='k-means++', verbose=0).fit(X_train)
    wcss.append(k_cluster.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(1,11) ,wcss )
plt.xticks(range(1,11))
plt.xlabel("WCSS")
plt.ylabel("Clusters")
```

```python
k_cluster = KMeans(n_clusters=3, init='k-means++', verbose=0).fit(X_train)
k_cluster.fit(X_train)
pred = k_cluster.fit_predict(X_train)
print(pred)
```

```python
plt.scatter(X_train[:,0] , X_train[:,1], c =pred )
```

```python
test_pred = k_cluster.predict(X_test)
print(test_pred)
```

```python
plt.scatter(X_test[:,0] , X_test[:,1], c =test_pred )
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_test, test_pred)
```

```python
from kneed import KneeLocator
k1 = KneeLocator(range(1,11), wcss, curve="convex", direction="decreasing")
k1.elbow
```

```python
silhouette_score_lst = []
for i in range(2, 11):
    k_cluster = KMeans(n_clusters=i, init='k-means++', verbose=0)
    k_cluster.fit(X_train)
    score = silhouette_score(X_train, k_cluster.labels_)
    silhouette_score_lst.append(score)
```

```python
silhouette_score_lst
```

```python
plt.plot(range(2,11) ,silhouette_score_lst )
plt.xticks(range(2,11))
plt.xlabel("silhouette_score")
plt.ylabel("Clusters")
```
