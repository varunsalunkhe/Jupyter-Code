```python
from keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.utils import load_img, img_to_array
```

```python
model= VGG16()
```

```python
model.summary()
```

```python
from keras.models import Model
sub= Model(inputs= model.input, outputs= model.layers[6].output)
```

```python
sub.summary()
```

```python
madara= load_img("Uchiha-Clan.jpg", target_size=(224,224))
madara
```

```python
madara= preprocess_input(img_to_array(madara))
madara.shape
```

```python
from numpy import expand_dims
madara= expand_dims(madara, axis=0)
```

```python
madara.shape
```

```python
samp = sub.predict(madara)
```

```python
from matplotlib import pyplot as plt
ax=plt.subplot(1,2,1)
ax.imshow(samp[0,:,:,4])
```

```python
plt.figure(figsize=(10,10))
b=1
for i in range(10):
    for i in range(10):
        ax=plt.subplot(10,10,b)
        ax.imshow(samp[0,:,:,b-10])
        b=b+1
        
```
