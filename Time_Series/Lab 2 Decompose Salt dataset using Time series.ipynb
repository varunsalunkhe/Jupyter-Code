# In class submission: Tahoe salt using Python TS decomposition

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose

df=pd.read_csv("salt.csv")
 
df.head()
```

```python
df.shape
```

```python
df.plot()
plt.show()
```

```python
result=seasonal_decompose(df['Demand'], model='multiplicable', period=4)
```

```python
result.seasonal.plot()
```

```python
result.trend.plot()
```

```python
result.plot()
plt.show()
```

# # In class submission: Random number based  Python TS decomposition for practise

```python
df = pd.DataFrame(np.random.randn(500), columns=["Demand"]).cumsum()
print(len(df))
df["Period"] = pd.Series(list(range(len(df))))
df.plot(x="Period", y="Demand")
```

```python
result=seasonal_decompose(df['Demand'], model='additive', period=100)
```

```python
result.seasonal.plot()
```

```python
result.trend.plot()
```
