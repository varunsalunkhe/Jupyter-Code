```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
%matplotlib inline
```

# count plot

```python
plt.figure(figsize=[15,5])
sns.countplot(x='sepal_length',data=iris,palette='rocket')
```

```python
plt.figure(figsize=[5,15])
sns.countplot(y='sepal_length',data=iris,palette='rocket')
```

```python
style.use("ggplot")
plt.figure(figsize=[15,5])
sns.countplot(x='sepal_length',hue='species',data=iris,palette='rocket')
plt.legend()
```

# displot

```python
style.use("ggplot")
sns.displot(iris.sepal_length,bins=20,edgecolor='k',palette='set1')
```

```python
plt.figure(figsize=[8,8])
sns.heatmap(iris.corr(),cbar=True,linewidth=2)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,color="blue")
```

```python
# plt.style.use("Seaborn_whitegrid")
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris,color="blue",size="sepal_length",alpha=0.5,markers=["^","*","o"])
```

```python
style.use("ggplot")
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris,color="blue",size="species",alpha=0.9,markers=["^","*","o"])
```

```python
style.use("ggplot")
sns.pairplot(iris)
```

```python
style.use("ggplot")
sns.pairplot(iris,hue="species",height=2.4,palette='rocket')
```

# lmplot

```python
sns.lmplot(x='sepal_length',y='petal_length',hue='species',data=iris,palette='rocket',markers=["^","*","o"])
```
